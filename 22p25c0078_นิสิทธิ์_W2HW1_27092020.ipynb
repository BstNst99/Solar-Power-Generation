```
def createdfSerie(df12,howJoin,daystart=1,dayneed = 3,y_field='TOTAL_YIELD',sec_div=3600):
  dfserie =df12[[y_field,'DATE_TIME','SOURCE_KEY']].copy().rename(columns={y_field: "Y_"+y_field})
  dfserie['DATE_TIME'] = dfserie['DATE_TIME'] - np.timedelta64(daystart,'D')
  #print(1,y_field)
  for shift in range(dayneed):
    dfserie=pd.merge(dfserie,df12, on=["DATE_TIME", "SOURCE_KEY"], how=howJoin,suffixes=('',str(daystart+shift)))
    dfserie['DATE_TIME'] = dfserie['DATE_TIME'] - np.timedelta64(1,'D')

  #dfserie['DATE_TIME']= dfserie['DATE_TIME'] + np.timedelta64(daystart+dayneed,'D')
  dfserie['DATE_TIME']=pd.to_timedelta(dfserie['DATE_TIME'].dt.strftime('%H:%M:%S')).dt.total_seconds()/sec_div
  return dfserie
```

```
from sklearn.metrics import mean_squared_error,confusion_matrix
```

```
!pip install catboost

from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

xgb = XGBRegressor(n_estimators=50,
                       objective ='reg:squarederror',
                       learning_rate = 0.1,
                       colsample_bytree=0.6,
                       max_depth = 3,
                       min_child_weight = 6)
cb = CatBoostRegressor(iterations=50, learning_rate=0.16,  depth=4,  verbose=0)
lgb = LGBMRegressor()
rf = RandomForestRegressor(max_depth=3, random_state=0)
LR = LinearRegression()

Models = [xgb,cb,lgb,rf,LR]
```

```
def model(clf,x_train,y_train,x_test,y_test,fold):
  clf.fit(x_train,y_train)
  y_pred = clf.predict(x_test)
  print('**************************************')
  print('at Fold :{}'.format(str(fold)))
  print('**************************************')
  print('Train score : {}'.format(clf.score(x_train,y_train)))
  print('Test score : {}'.format(clf.score(x_test,y_test)))
  print('Root Mean Square Error : {}'.format(mean_squared_error(y_test,y_pred)**0.5))
  return clf,y_pred,clf.score(x_train,y_train),clf.score(x_test,y_test),mean_squared_error(y_test,y_pred)**0.5
```

```
def TestAllModal(train10f,test10f,Y_field='TOTAL_YIELD'):
  train_scores = []
  test_scores = []
  rmse_scores = []
  compare_scores = []

  for mod in Models:
    print('**************************************')
    print('Model : {}'.format(type(mod).__name__))
    print('**************************************')
    #train10f[0].head()
    for fold in range(folds):
      #x_train = train10f[fold].drop(['Y_'+Y_field,'DATE_TIME'],axis=1)
      x_train = train10f[fold].drop(['Y_'+Y_field],axis=1)
      y_train = train10f[fold]['Y_'+Y_field]
      #x_test = test10f[fold].drop(['Y_'+Y_field,'DATE_TIME'],axis=1)
      x_test = test10f[fold].drop(['Y_'+Y_field],axis=1)
      y_test = test10f[fold]['Y_'+Y_field]

      print('Train size : {}'.format(train10f[fold].shape))
      print('Test size : {}'.format(test10f[fold].shape))
      res_mod,y_pred,train_score,test_score,rmse_score = model(mod,x_train,y_train,x_test,y_test,fold)
      train_scores.append(train_score)
      test_scores.append(test_score)
      rmse_scores.append(rmse_score)
    print('**************************************')
    print('Average Train score: {}'.format(np.mean(train_scores)))
    print('Average Test score: {}'.format(np.mean(test_scores)))
    print('Average RMSE score: {}'.format(np.mean(rmse_scores)))
    compare_scores.append([type(mod).__name__,np.mean(train_scores),np.mean(test_scores),np.mean(rmse_scores)])

  return pd.DataFrame(compare_scores,columns=['Predictor','Training score','Testing score','RMSE score'])
```

```
def pipline(df12,daystart=1,dayneed = 3,howJoin='',y_field='TOTAL_YIELD',sec_div=3600,dropfield=''):
  
  dfserie=createdfSerie(df12,howJoin,daystart,dayneed,y_field,sec_div)

  if dropfield!='':
    dfserie=dfserie.drop([dropfield],axis=1)

  df10f = [dfserie[dfserie.index % folds == i].fillna(0) for i in range(folds)] 
  
  train10f=[df10f[i][:int(len(df10f[i])*0.9)] for i in range(folds)] 
  test10f =[df10f[i][int(len(df10f[i])*0.9):] for i in range(folds)] 
  summary=TestAllModal(train10f,test10f,y_field)
  return summary,dfserie,train10f
```
